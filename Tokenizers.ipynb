```python
# Install Hugging Face transformers
!pip install transformers


from transformers import pipeline

# Load a real sentiment analysis model
classifier = pipeline("sentiment-analysis")

# Test it
result = classifier("I love learning about AI!")
print(result)

sentences = [
    "I hate rainy days",
    "This is the best day of my life",
    "I am feeling okay I guess",
    "Machine learning is absolutely amazing",
    "म तिमीलाई माया गर्छु"
]

results = classifier(sentences)

for sentence, result in zip(sentences, results):
    label = result['label']
    score = round(result['score'] * 100, 2)
    print(f"{label} ({score}%) — {sentence}")

    # Try different types of pipelines
generator = pipeline("text-generation", model="gpt2")
result = generator("I am from Nepal and I love", max_length=30)
print(result[0]['generated_text'])



from transformers import AutoTokenizer

# Load a real tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

# Tokenize some text
text = "I am eating dal bhat in Kathmandu with Sambridhi and Bhabuk 😊 मोमो"
tokens = tokenizer.tokenize(text)
print("Text:", text)
print("Tokens:", tokens)
print("Number of tokens:", len(tokens))

token_ids = tokenizer.encode(text)
print("Token IDs:", token_ids)

# Mathematical equations
math_text = "y = mx + c"
math_tokens = tokenizer.tokenize(math_text)
print("Math equation:", math_tokens)

# Greek alphabets
greek_text = "alpha beta gamma delta"
greek_tokens = tokenizer.tokenize(greek_text)
print("Greek words:", greek_tokens)

# Actual Greek symbols
symbols_text = "α β γ δ"
symbol_tokens = tokenizer.tokenize(symbols_text)
print("Greek symbols:", symbol_tokens)

# Mathematical formula
formula_text = "E = mc²"
formula_tokens = tokenizer.tokenize(formula_text)
print("Einstein formula:", formula_tokens)

#Python code
code_text = "for i in range(10): print(i)"
code_tokens = tokenizer.tokenize(code_text)
print("Python code:", code_tokens)

# And numbers
number_text = "I have 1000 rupees and 50 paisa"
number_tokens = tokenizer.tokenize(number_text)
print("Numbers:", number_tokens)

# What about your own name vs common English names?
names = ["Sambridhi", "John", "Emma", "Bhabuk", "Priya"]

for name in names:
    tokens = tokenizer.tokenize(name)
    pieces = len(tokens)
    print(f"{name:12} → {tokens} ({pieces} piece{'s' if pieces > 1 else ''})")
```
